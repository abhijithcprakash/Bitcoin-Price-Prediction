# Normalizing the  Data 

```python
def normalize(df):
  result = df.copy()
  for feature_name in df.columns:
    max_value = df[feature_name].max()
    min_value = df[feature_name].min()
    result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
  return result
```

```python
bit_c=normalize(bitcoin)
```

```python
bit_c.head()
```

```python
Xx = bit_c.iloc[:, 0:-1]
yy = bit_c.target
```

```python
X_trainn, X_testt, y_trainn, y_testt = train_test_split(Xx, yy, test_size=0.3, random_state=1)
```

```python
y_predd = model2.predict(X_testt)
print('Accuracy: %.3f' % accuracy_score(y_testt, y_predd))
```

```python
def predictt(X_train, X_test, y_train, y_test, model):
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    preds = pd.Series(preds, index=X_test.index, name="prediction")
    combined = pd.concat([y_test, preds], axis=1)
    return combined
```

```python
def backtestt(data, model, start=1095, step=150):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        X_train = X.iloc[0:i].copy()
        X_test = X.iloc[i:(i+step)].copy()
        y_train= y.iloc[0:i].copy()
        y_test = y.iloc[i:(i+step)].copy()
        predictions = predictt(X_train, X_test, y_train, y_test, model)
        all_predictions.append(predictions)
    
    return pd.concat(all_predictions)
```

Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train =sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
# from keras.layers import LeakyRelu,PRelu,Elu
```

initialize ann

```python
classifier=Sequential()
```

Adding the input layer and first hidden layer

```python
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu',input_dim=9))
```

second hidden layer

```python
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))
```

Output layer

```python
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
classifier.compile(optimizer="adam",loss='binary_crossentropy',metrics=['accuracy'])
```

```python
classifier.summary()
```

```python
model_history=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=500)
```

```python
print(model_history.history.keys())
# summarize history for accuracy
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# summarize history for loss
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
